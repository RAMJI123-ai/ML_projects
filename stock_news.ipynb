```python
# Fetches all news

# 📥 Import libraries
import feedparser

# 🔍 Google News RSS feed for Indian companies (filtered by relevant terms)
google_rss_url = "https://news.google.com/rss/search?q=solar+stocks+OR+renewable+energy+OR+PLI+scheme+India+NSE+BSE"

# 📰 Bloomberg RSS feed for Indian stocks (filtered by relevant terms)
bloomberg_rss_url = "https://www.bloomberg.com/feeds/bigdata/rss/news"

# 🎯 Keywords to track relevant Indian news
impact_keywords = [
    "solar", "renewable", "green energy", "PLI scheme", "stock rally",
    "MNRE", "government policy", "subsidy", "earnings", "Q4", "Q1", "results", "exports",
    "India", "NSE", "BSE", "stock market", "Indian market"
]

# 🧾 Companies to track
companies = ["Waaree", "Adani Green", "Tata Power", "Borosph", "ReNew", "JSW Energy", "NTPC", "Reliance", "L&T"]

# 🔁 Function to fetch and filter news
def fetch_news():
    print("🔍 Checking for relevant Indian companies' news...\n")

    # Parse Google News Feed
    google_feed = feedparser.parse(google_rss_url)
    # Parse Bloomberg Feed
    bloomberg_feed = feedparser.parse(bloomberg_rss_url)

    found = False

    # Fetch news from Google News
    print("📰 Google News:\n")
    for entry in google_feed.entries:
        title = entry.title
        link = entry.link

        # Filter news: Check for relevant keywords or company names in the title
        if (any(word.lower() in title.lower() for word in impact_keywords) or
            any(company.lower() in title.lower() for company in companies)):
            print(f"✅ {title}\n🔗 {link}\n")
            found = True

    # Fetch news from Bloomberg
    print("📰 Bloomberg News:\n")
    for entry in bloomberg_feed.entries:
        title = entry.title
        link = entry.link

        # Filter news: Check for relevant keywords or company names in the title
        if (any(word.lower() in title.lower() for word in impact_keywords) or
            any(company.lower() in title.lower() for company in companies)):
            print(f"✅ {title}\n🔗 {link}\n")
            found = True

    if not found:
        print("ℹ️ No relevant news found right now.")

# ▶️ Run the script
fetch_news()
```

```python
import feedparser
import pandas as pd
import re
from collections import Counter
import os

# Economic Times Market RSS feed
et_rss = "https://economictimes.indiatimes.com/markets/rssfeeds/1977021501.cms"

# Bullish sentiment indicators
bullish_keywords = [
    "rallies", "surges", "jumps", "soars", "hits new high", "strong", "beats",
    "beats estimates", "up", "outperforms", "spikes", "bullish",
    "positive results", "Q4 results", "profit"
]

# Download NSE stock list if not present
nse_url = "https://archives.nseindia.com/content/equities/EQUITY_L.csv"
nse_file = "nse_stocks.csv"

if not os.path.exists(nse_file):
    print("📥 Downloading NSE stock list...")
    import requests
    r = requests.get(nse_url)
    with open(nse_file, 'wb') as f:
        f.write(r.content)

# Load Indian stock names
nse_df = pd.read_csv(nse_file)
nse_names = set(nse_df['NAME OF COMPANY'].str.upper().str.strip())

# Company name-like pattern
company_pattern = re.compile(r'\b([A-Z][a-zA-Z&.\-]{2,}(?:\s+[A-Z][a-zA-Z&.\-]{2,})*)\b')

def get_top_bullish_indian_stocks(limit=3):
    print("\n🔍 Scanning Economic Times for bullish Indian stock news...\n")

    feed = feedparser.parse(et_rss)
    score_counter = Counter()

    for entry in feed.entries:
        title = entry.title
        title_lower = title.lower()

        if any(word in title_lower for word in bullish_keywords):
            candidates = company_pattern.findall(title.upper())

            for name in candidates:
                clean_name = name.strip().upper()
                if clean_name in nse_names:
                    score_counter[clean_name] += 1

    if score_counter:
        print("📈 Top 3 Bullish Indian Stocks (From ET Headlines):\n")
        for name, score in score_counter.most_common(limit):
            print(f"✅ {name.title()} — Score: {score}")
    else:
        print("ℹ️ No bullish news found for Indian stocks.")

# ▶️ Run it
get_top_bullish_indian_stocks()
```
