```python
# ðŸš€ NSE 5%+ Gainers Screener with News Headlines (for Indian Stock Market)

# âœ… Step 2: Import Modules
import yfinance as yf
import pandas as pd
from datetime import datetime
import requests
from bs4 import BeautifulSoup
from googlesearch import search

# âœ… Step 3: Define Function to Get Top 5% Gainers from NSE
def get_top_gainers(min_gain=5.0):
    nifty_stocks = [
        "RELIANCE.NS", "TCS.NS", "INFY.NS", "WIPRO.NS", "HDFCBANK.NS",
        "ICICIBANK.NS", "SBIN.NS", "HINDUNILVR.NS", "TATAMOTORS.NS", "DEEPAKNTR.NS"
    ]
    top_gainers = []
    for ticker in nifty_stocks:
        df = yf.download(ticker, period="1d", interval="5m", progress=False)
        if df.empty:
            continue
        df["% Change"] = ((df["Close"] - df["Open"]) / df["Open"]) * 100
        max_change = df["% Change"].max()
        if max_change > min_gain:
            top_gainers.append({
                "Stock": ticker,
                "Max % Gain": round(max_change, 2),
                "Latest Price": df["Close"].iloc[-1]
            })
    return pd.DataFrame(top_gainers)

# âœ… Step 4: Sector Mapping (You can expand this)
sector_map = {
    "RELIANCE.NS": "Oil & Gas",
    "TCS.NS": "IT",
    "INFY.NS": "IT",
    "WIPRO.NS": "IT",
    "HDFCBANK.NS": "Banking",
    "ICICIBANK.NS": "Banking",
    "SBIN.NS": "Banking",
    "HINDUNILVR.NS": "FMCG",
    "TATAMOTORS.NS": "Auto",
    "DEEPAKNTR.NS": "Chemicals"
}

# âœ… Step 5: News Scraper using Google Search
def get_latest_news(stock_name):
    query = f"{stock_name} stock news site:moneycontrol.com"
    try:
        for url in search(query, num_results=1):
            res = requests.get(url)
            soup = BeautifulSoup(res.text, 'html.parser')
            headline = soup.title.string.strip()
            return headline
    except:
        return "No news found"


# âœ… Step 6: Build Screener
def build_screener():
    gainers = get_top_gainers()
    # Check if gainers DataFrame is empty before proceeding
    if gainers.empty:
        print("No stocks found with a gain greater than 5%.")
        return pd.DataFrame()  # Return empty DataFrame to avoid error
    gainers["Sector"] = gainers["Stock"].map(sector_map)
    gainers["Recent News"] = gainers["Stock"].apply(lambda x: get_latest_news(x.replace(".NS", "")))
    return gainers

# âœ… Run the Screener
screener_df = build_screener()
screener_df
```
