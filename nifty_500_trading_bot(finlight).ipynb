```python
import yfinance as yf
import pandas as pd
import time
from datetime import datetime, timezone

# Parameters
TICKER = "^NSEI"
PERIOD = "5d"
INTERVAL = "5m"
SMA_WINDOW = 20
SLEEP_SECONDS = 180

# History DataFrame
signal_history = pd.DataFrame(columns=["Time", "Price", "Signal"])

def fetch_data():
    data = yf.download(
        tickers=TICKER,
        period=PERIOD,
        interval=INTERVAL,
        progress=False,
        threads=False,
        auto_adjust=True
    )
    data['SMA'] = data['Close'].rolling(window=SMA_WINDOW).mean()
    data = data.dropna()
    return data

def generate_signal(df):
    close = float(df['Close'].iloc[-1])
    sma = float(df['SMA'].iloc[-1])

    if close > sma:
        return "BUY"
    elif close < sma:
        return "SELL"
    else:
        return "HOLD"

# Main loop
while True:
    try:
        df = fetch_data()
        signal = generate_signal(df)
        last_price = float(df['Close'].iloc[-1])
        loop_time = datetime.now(timezone.utc)

        # Append to history
        signal_history = pd.concat([signal_history, pd.DataFrame([{
            "Time": loop_time,
            "Price": last_price,
            "Signal": signal
        }])], ignore_index=True)

        # Print
        print(f"\n=== Loop at {loop_time} | Scanning {TICKER} ===")
        print(f"Signal: {signal} | Last Price: {last_price:.2f}")
        print("\n--- Recent Signal History ---")
        print(signal_history.tail(5).to_string(index=False))

    except Exception as e:
        print(f"Error fetching/signaling {TICKER}: {e}")

    time.sleep(SLEEP_SECONDS)
```
